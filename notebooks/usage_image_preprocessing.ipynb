# 1. How to use customed features?

## 1.1. Sharpen images

```python
image_path = '../data/development/2025.png'
image = cv2.imread(image_path)

plt.imshow(image)
```

```python
sharpened_image = processor.sharpen_image(
    image, 
    sharpen_factor=1.5
)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(sharpened_image)
ax[1].set_title('Image with Adaptive Median Filter')
ax[1].axis('off')

plt.show()
```

## 1.2. Filter images

```python
filtered_image = processor.filter_noise(
    input_image=image,
    sigma=1
)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(filtered_image)
ax[1].set_title('Image with Adaptive Median Filter')
ax[1].axis('off')

plt.show()
```
